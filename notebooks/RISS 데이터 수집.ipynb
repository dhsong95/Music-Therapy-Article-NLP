## RISS 데이터 수집 (Crawling)

* 수집 대상: 한국음악치료학회지

```python
# Step 1. 필요한 라이브러리 모듈 설치
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas
import re
```

```python
# Step 2. 기본 정보 설정
base_url = 'http://www.riss.kr/search/detail/DetailView.do?p_mat_type=3a11008f85f7c51d&control_no=2a75a048a6856c25ffe0bdc3ef48d419'
```

* 크롤링 과정
    * 개별 논문에 대한 접근 정보(url) 구하기 (vol, no, id, url)
    * 개별 논문에 대한 메타 정보 구히가 (제목, 저자, 권호 사항(연도, vol, no), 작성 언어, 등재 정보, 자료 형태, 수록면, 피인용회수, 국문 초록)
* 크롤링 시 데이터가 없는 경우에 대한 처리 필요

```python
# Step 3. 드라이버 설정
driver = webdriver.Chrome('./driver/chromedriver')
driver.get(base_url)
time.sleep(2)
```

```python
# Step 4. 개별 논문 접근 정보 저장
url_list = []
title_list = []

# 모든 정보 클릭해서 활성화
for idx in range(1, 22):
    year = driver.find_element_by_xpath('//*[@id="divContent"]/div[1]/div/div[2]/ul/li[' + str(idx) + ']/a')
    driver.execute_script('arguments[0].click()', year)
    time.sleep(0.5)
    
    no_list = year.find_elements_by_xpath('//*[@id="divContent"]/div[1]/div/div[2]/ul/li[' + str(idx) + ']/ul/li')
    
    for no in no_list:
        driver.execute_script('arguments[0].click()', no.find_element_by_xpath('a'))
        time.sleep(0.5)
      
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        elements = soup.select('#soptionview > div > div.srchResultW.bd > div.srchResultListW > ul > li')
        for element in elements:
            title = element.find('div', 'cont').find('p', 'title').get_text().strip()
            url = element.find('div', 'cont').find('p', 'title').find('a')['href']
            print('{:s}\t{:s}'.format(title, url))
            title_list.append(title)
            url_list.append(url)
        

title_url_df = pandas.DataFrame()
title_url_df['title'] = title_list
title_url_df['url'] = url_list
title_url_df.to_csv('./data/urls', encoding='utf-8', index=False)
```

```python
# Step 5. 개별 논문 메타 정보 저장
titles = pandas.read_csv('./data/urls')['title']
urls = pandas.read_csv('./data/urls')['url']

title_list = []
author_list = []
org_list = []
name_list = []
volno_list = []
year_list = []
lang_list = []
keyword_list = []
kdc_list = []
kci_list = []
media_list = []
page_list = []
citation_list = []
link_list = []
abstract_list = []
loc_list = []

for title, url in zip(titles, urls):
    url = 'http://www.riss.kr' + url
    driver.get(url)
    time.sleep(2)
    
    mores = driver.find_elements_by_class_name('moreView')
    for more in mores:
        driver.execute_script('arguments[0].click()', more)
        time.sleep(0.5)
    time.sleep(1)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    author = 'MV'
    org = '한국음악치료학회'
    name = '한국음악치료학회지'
    volno = 'MV'
    year = 'MV'
    lang = 'MV'
    keyword = 'MV'
    kdc = 'MV'
    kci = 'MV'
    media = '학술저널'
    page = 'MV'
    citation = 'MV'
    link = 'MV'
    abstract = 'MV'
    loc = 'MV'
    
    elements = soup.select('#soptionview > div > div.thesisInfo > div.infoDetail > div.infoDetailL > ul > li')
    for element in elements:
        key = element.find('span', 'strong').get_text().strip()
        value = element.find('div').get_text().strip()
        
        if key == '저자':
            author = value
        elif key == '발행기관':
            org = value
        elif key == '학술지명':
            name = value
        elif key == '권호사항':
            volno = ' '.join(value.split())
        elif key == '발행연도':
            year = value
        elif key == '작성언어':
            lang = value
        elif key == '주제어':
            keyword = ' '.join(value.split())
        elif key == 'KDC':
            kdc = value
        elif key == '등재정보':
            kci = value
        elif key == '자료형태':
            media = value
        elif key == '수록면':
            page = value
        elif key == 'KCI 피인용지수' or key == 'KCI 피인용횟수':
            citation = value
        elif key == '제공처':
            link = value
        elif key == '소장기관':
            loc = value
        else:
            print('*********New Key: {:s}*********'.format(key))

    additionals = soup.find_all('div', 'innerCont')
    for addition in additionals:
        if addition.find('h3', 'tit').get_text().strip() == '부가정보':
            contents = addition.select('div.content > div')
            for content in contents:
                if content.find('p', 'title').get_text().strip() == '국문 초록 (Abstract)':
                    abstract = content.find_all('div', 'text')[1].find('p').get_text().strip()
                    break
        
    print('{:s}\t{:s}\t{:s}\n\t{:s}'.format(title, volno, keyword, abstract))
    title_list.append(title)
    author_list.append(author)
    org_list.append(org)
    name_list.append(name)
    volno_list.append(volno)
    year_list.append(year)
    lang_list.append(lang)
    keyword_list.append(keyword)
    kdc_list.append(kdc)
    kci_list.append(kci)
    media_list.append(media)
    page_list.append(page)
    citation_list.append(citation)
    link_list.append(link)
    abstract_list.append(abstract)  

article_df = pandas.DataFrame()
article_df['title'] = title_list
article_df['author'] = author_list
article_df['organization'] = org_list
article_df['name'] = name_list
article_df['volume no'] = volno_list
article_df['year'] = year_list
article_df['language'] = lang_list
article_df['keyword'] = keyword_list
article_df['KDC'] = kdc_list
article_df['KCI'] = kci_list
article_df['media'] = media_list
article_df['page'] = page_list
article_df['citation'] = citation_list
article_df['link'] = link_list
article_df['abstract'] = abstract_list

article_df.to_csv('./data/articles.csv', encoding='utf-8', index=False)
```

```python
# Step 6. 데이터 정리
article_df = pandas.read_csv('./data/articles.csv', encoding='utf-8')

title_list = article_df['title']
abstract_list = article_df['abstract']
keyword_list = article_df['keyword']
year_list = article_df['year']
author_list = []
vol_list = []
no_list = []

for author in article_df['author']:
    author = ' '.join(author.split())
    author_list.append(author)

for volno in article_df['volume no']:
    vol = volno.split(' ')[0]
    no = volno.split(' ')[1]
    vol = vol.replace('Vol.', '')
    no = vol.replace('No.', '')
    vol_list.append(vol)
    no_list.append(no)

article_essential_df = pandas.DataFrame()
article_essential_df['title'] = title_list
article_essential_df['author'] = author_list
article_essential_df['year'] = year_list
article_essential_df['vol'] = vol_list
article_essential_df['no'] = no_list
article_essential_df['keyword'] = keyword_list
article_essential_df['abstract'] = abstract_list

article_essential_df.to_csv('./data/articles_essential.csv', encoding='utf-8', index=False)
```
